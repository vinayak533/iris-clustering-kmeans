
# ðŸŒ¸ Iris Dataset Clustering with KMeans

This project demonstrates unsupervised machine learning using the **KMeans clustering algorithm** on the famous **Iris flower dataset**. 

### Objectives:
- Explore the structure of the dataset using scatter plots.
- Apply KMeans clustering on both original and scaled data.
- Use the **Elbow Method** to find the optimal number of clusters.
- Visually compare clustering results with the actual species labels.

ðŸ”§ Tools used: Python, Pandas, Matplotlib, Seaborn, Scikit-learn




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data = pd.read_csv('iris_dataset.csv')

print(data)
```

```python

plt.scatter(data['sepal_length'],data['sepal_width'])

plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

```python

x = data.copy()


kmeans = KMeans(2)

kmeans.fit(x)
```

```python

clusters = data.copy()

clusters['cluster_pred']=kmeans.fit_predict(x)

```

```python

plt.scatter(clusters['sepal_length'], clusters['sepal_width'], c= clusters ['cluster_pred'], cmap = 'rainbow')
```

```python

from sklearn import preprocessing


x_scaled = preprocessing.scale(data)
x_scaled
```

```python

kmeans_scaled = KMeans(2)

kmeans_scaled.fit(x_scaled)
```

```python

clusters_scaled = data.copy()

clusters_scaled['cluster_pred']=kmeans_scaled.fit_predict(x_scaled)
```

```python

plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['cluster_pred'], cmap = 'rainbow')
```

```python
wcss = []

cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
